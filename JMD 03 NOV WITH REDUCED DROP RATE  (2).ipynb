```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Function to plot the boxplot for a single feature with Min-Max normalization
def plot_minmax_boxplot(data, feature_index, title):
    feature_data = data[:, feature_index]
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=feature_data)
    plt.title(title)

    # Calculate Q1, Q3, IQR, and whiskers
    Q1, Q3 = np.percentile(feature_data, [25, 75])
    IQR = Q3 - Q1
    lower_whisker = Q1 - 1.5 * IQR
    upper_whisker = Q3 + 1.5 * IQR

    # Plot the whiskers and outliers
    plt.axvline(Q1, color='green', linestyle='--', label='Q1 (25th Percentile)')
    plt.axvline(Q3, color='blue', linestyle='--', label='Q3 (75th Percentile)')
    plt.axvline(lower_whisker, color='red', linestyle=':', label='Lower Whisker')
    plt.axvline(upper_whisker, color='red', linestyle=':', label='Upper Whisker')

    # Identify and plot outliers
    outliers = feature_data[(feature_data < lower_whisker) | (feature_data > upper_whisker)]
    plt.scatter(outliers, np.zeros_like(outliers), color='purple', zorder=5, label='Outliers')

    plt.legend()
    plt.show()

# Apply function to the first feature in Min-Max Normalized data
plot_minmax_boxplot(train_data_minmax, feature_index=0, title="Min-Max Normalized Feature 1 (Training Set)")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.model_selection import KFold
import tensorflow as tf
from tensorflow.keras import layers, regularizers
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# Ensure we are using the GPU
with tf.device('/GPU:0'):  # Use GPU explicitly
    print("Training on GPU")

# Define Autoencoder Building Function with L2 Regularization and Dropout
def build_autoencoder(latent_dim, n_layers, dropout_rate, l2_penalty):
    inputs = tf.keras.Input(shape=(train_data_minmax.shape[1],))
    x = inputs

    # Encoder with L2 regularization and dropout
    for _ in range(n_layers // 2):
        x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(l2_penalty))(x)
        if dropout_rate:
            x = layers.Dropout(dropout_rate)(x)

    # Bottleneck (latent space)
    encoded = layers.Dense(latent_dim, activation='relu')(x)

    # Decoder
    x = encoded
    for _ in range(n_layers // 2):
        x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(l2_penalty))(x)
        if dropout_rate:
            x = layers.Dropout(dropout_rate)(x)

    # Changed output activation to 'linear' for regression tasks
    decoded = layers.Dense(train_data_minmax.shape[1], activation='linear')(x)
    autoencoder = tf.keras.Model(inputs, decoded)

    return autoencoder

# Compute Reconstruction Accuracy Function
def compute_reconstruction_accuracy(original_data, reconstructed_data):
    mse = np.mean(np.square(original_data - reconstructed_data), axis=1)
    threshold = 0.01  # Adjustable threshold for accuracy measurement
    accuracy = np.mean(mse < threshold)
    return accuracy

# Plotting Function for Results
def plot_results(accuracies, losses):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(range(1, len(accuracies) + 1), accuracies, marker='o', color='b', label='Accuracy')
    plt.title("Average Reconstruction Accuracy per Autoencoder")
    plt.xlabel("Autoencoder")
    plt.ylabel("Accuracy")
    plt.xticks(range(1, len(accuracies) + 1))
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(range(1, len(losses) + 1), losses, marker='o', color='r', label='Loss')
    plt.title("Average Reconstruction Loss per Autoencoder")
    plt.xlabel("Autoencoder")
    plt.ylabel("Loss")
    plt.xticks(range(1, len(losses) + 1))
    plt.legend()
    plt.tight_layout()
    plt.show()

# K-Fold Cross-Validation and Model Training Function
def train_and_evaluate_autoencoders(data, latent_dim, dropout_rates, n_layers, l2_penalty, learning_rate, batch_size, epochs):
    autoencoders = []
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    all_fold_accuracies = []
    all_fold_losses = []

    for i, (layers, dropout_rate) in enumerate(zip(n_layers, dropout_rates), start=1):
        print(f"\nTraining Autoencoder AE{i} with {layers} layers and Dropout Rate: {dropout_rate}")

        fold_accuracies = []
        fold_losses = []

        # Perform 5-fold cross-validation
        for fold, (train_idx, val_idx) in enumerate(kf.split(data)):
            print(f"\nStarting Fold {fold + 1} for AE{i}")

            train_data = data[train_idx]
            val_data = data[val_idx]

            # Build and compile model for each fold
            autoencoder = build_autoencoder(latent_dim, layers, dropout_rate, l2_penalty)
            autoencoder.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate), loss='mean_squared_error')

            # Early Stopping callback
            early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

            # Train the model
            history = autoencoder.fit(
                train_data, train_data,
                validation_data=(val_data, val_data),
                epochs=epochs,
                batch_size=batch_size,
                callbacks=[early_stopping],
                verbose=1  # Set verbose to 1 for progress updates
            )

            # Check if early stopping was triggered
            if len(history.history['val_loss']) < epochs:
                print(f"EARLY STOPPING TRIGGERED AT Epoch {len(history.history['val_loss'])}")

            # Evaluate on validation set
            val_loss = autoencoder.evaluate(val_data, val_data, verbose=0)
            fold_losses.append(val_loss)

            # Compute reconstruction accuracy on validation data
            reconstructed_val_data = autoencoder.predict(val_data, verbose=0)
            reconstruction_accuracy = compute_reconstruction_accuracy(val_data, reconstructed_val_data)
            fold_accuracies.append(reconstruction_accuracy)

            # Print fold results
            print(f"FOLD {fold + 1} AE{i} Reconstruction Accuracy: {reconstruction_accuracy:.4f}")
            print(f"FOLD {fold + 1} AE{i} Reconstruction Loss: {val_loss:.4f}")

        # Store average results for each autoencoder
        avg_accuracy = np.mean(fold_accuracies)
        avg_loss = np.mean(fold_losses)
        all_fold_accuracies.append(avg_accuracy)
        all_fold_losses.append(avg_loss)

        print(f"\nAE{i} - Average Cross-Validated Reconstruction Accuracy: {avg_accuracy:.4f}")
        print(f"AE{i} - Average Cross-Validated Reconstruction Loss: {avg_loss:.4f}\n")
        autoencoders.append(autoencoder)

    # Display results after each autoencoder completes
    plot_results(all_fold_accuracies, all_fold_losses)
    return autoencoders

# Set parameters for training
latent_dim = 30
dropout_rates = [0.2, 0.25, 0.3, 0.35, 0.4, 0.45]  # Increased dropout rates for overfitting reduction
n_layers = [6, 8, 8, 8, 8, 10]
l2_penalty = 0.0001  # Reduced regularization penalty to avoid over-penalizing
learning_rate = 0.001  # Slightly increased learning rate
batch_size = 64
epochs = 20

# Train and evaluate autoencoders with updated function
autoencoders = train_and_evaluate_autoencoders(
    data=train_data_minmax,
    latent_dim=latent_dim,
    dropout_rates=dropout_rates,
    n_layers=n_layers,
    l2_penalty=l2_penalty,
    learning_rate=learning_rate,
    batch_size=batch_size,
    epochs=epochs
)

# Output list of trained autoencoders
print("Trained Autoencoders:", autoencoders)
```

```python
# Step 5 - Generate Encoded Representations
# Generate encoded representations for training and test data using all autoencoders
encoded_train = [ae.predict(train_data_minmax) for ae in autoencoders]
encoded_test = [ae.predict(test_data_minmax) for ae in autoencoders]

```

```python
# Lists to store RMSE and accuracy values for training and test sets
train_rmse_list = []
test_rmse_list = []
train_accuracy_list = []  # Added to store training accuracy
test_accuracy_list = []   # Added to store test accuracy

# Initialize the Decision Tree Classifier
dt_model = DecisionTreeClassifier()

# Loop through each autoencoder's encoded data
for i, encoded_data in enumerate(encoded_train, start=1):
    # Fit the decision tree model to the encoded training data
    dt_model.fit(encoded_data, train_data_labels)  # Assuming `train_data_labels` and `test_data_labels` are defined

    # Training set predictions and accuracy
    train_predictions = dt_model.predict(encoded_data)
    train_rmse = mean_squared_error(train_data_labels, train_predictions, squared=False)
    train_accuracy = accuracy_score(train_data_labels, train_predictions)  # Calculate and store training accuracy
    train_rmse_list.append(train_rmse)
    train_accuracy_list.append(train_accuracy)  # Append training accuracy to the list

    # Test set predictions and accuracy
    encoded_test_data = encoded_test[i-1]
    test_predictions = dt_model.predict(encoded_test_data)
    test_rmse = mean_squared_error(test_data_labels, test_predictions, squared=False)
    test_accuracy = accuracy_score(test_data_labels, test_predictions)  # Calculate and store test accuracy
    test_rmse_list.append(test_rmse)
    test_accuracy_list.append(test_accuracy)  # Append test accuracy to the list

    # Print results for current autoencoder
    print(f"RMSE TRAIN for AE{i} Encoded Data - {train_rmse:.4f}")
    print(f"RMSE TEST for AE{i} Encoded Data - {test_rmse:.4f}")
    print(f"Decision Tree Training Accuracy for AE{i}: {train_accuracy:.4f}")  # Print training accuracy
    print(f"Decision Tree Test Accuracy for AE{i}: {test_accuracy:.4f}")  # Print test accuracy
```
